```python
# Import the dependecies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy import column, Integer, String, Float
```

```python
#Read the netflix data and the study results
netflix_path = "netflix_titles.csv"
netflix_data = pd.read_csv(netflix_path)
netflix_data.head

```

```python
movies_netflix = netflix_data.loc[netflix_data["type"]=="Movie"]
movies_netflix.head()
```

```python
TV_Shows_netflix = netflix_data.loc[netflix_data["type"]=="TV Show"]
TV_Shows_netflix.head()
```

```python
movies_netflix.to_csv('movies.csv', index = False)
```

```python
TV_Shows_netflix.to_csv('TVShows.csv', index = False)
```

```python
# Make a connection 
conn = sqlalchemy.create_engine('postgresql://localhost:5432/Netflix_content?user=postgres&password=password')
```

```python
movies_netflix.to_sql("movies", con = conn, if_exists = 'replace', index = False)
```

```python
TV_Shows_netflix.to_sql("tv_shows", con = conn, if_exists = 'replace', index = False)
```

```python
movies_netflix_table = pd.read_sql(sql = "select * from movies", con=conn)
movies_netflix_table.head(50)
```

```python
TV_Shows_netflix_table = pd.read_sql(sql = "select * from tv_shows", con=conn)
TV_Shows_netflix_table.head(50)
```
